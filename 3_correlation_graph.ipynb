```r
# 自定义柱状图（空心蓝色）
my_bar <- function(data, mapping, ...) {
  ggplot(data, mapping) +
    geom_histogram(color = "white",fill="blue", bins = 20,size=0.5)
}

# 自定义散点图（实心红色点，黑色拟合线）
my_scatter <- function(data, mapping, ...) {
  ggplot(data, mapping) +
    geom_point(color = "red", alpha = 0.3, size = 1.5) +
    geom_smooth(method = "lm", color = "black", ...)
}

# 绘制相关性统计图
p <- ggpairs(
  data,
  lower = list(continuous = wrap(my_scatter)),
  diag = list(continuous = wrap(my_bar)),
  upper = list(continuous = wrap("cor", size = 5))
)

# 添加标题
p <- p + ggtitle("Scatterplot matrix for DTW, HAND, and SWI")

# 显示图表
print(p)
```
