```python
import random
import pandas as pd
import matplotlib.pyplot as plt
#解决负号无法正常显示的问题
plt.rcParams['axes.unicode_minus']=False
# 让图表直接在Jupyter Notebook中展示出来
%matplotlib inline
# #矢量图设置
%config InlineBackend.figure_format='svg'
# fonts
font1 = {"family": 'Times New Roman', "style": "normal",
        "variant": "small-caps", "weight": "medium", "size": 16}
font2 = {"family": 'Times New Roman', "style": "normal",
        "variant": "small-caps", "size": 12}
font3 = {"family": 'Times New Roman', "style": "normal",
        "variant": "small-caps", "weight": "bold", "size": 18}
font4 = {"family": 'Times New Roman', "style": "normal",
        "variant": "small-caps", "weight": "medium", "size": 14}
```

#### AUC Bar

```python
# AUC
x=['UB 2500','UB 5000','UB 10000',"BR 5000","TGB 5000","BTH 5000"]
y=[0.931,0.94,0.934,0.954,0.919,0.914]
plt.figure(figsize=(6, 8))
plt.bar(x,y,facecolor='c', alpha=0.75,edgecolor = 'k')
# 删除上方和右边的边框线，对下方和左边的边框线加粗
ax = plt.gca()
# ax.spines['top'].set_color('none')
# ax.spines['right'].set_color('none')
bwith = 2  # 边框宽度设置为2
ax.spines['bottom'].set_linewidth(bwith)
ax.spines['left'].set_linewidth(bwith)
ax.spines['top'].set_linewidth(bwith)
ax.spines['right'].set_linewidth(bwith)
# xy轴刻度与标签
plt.xticks(fontproperties='Palatino Linotype', size=12.5)
plt.yticks(fontproperties='Palatino Linotype', size=12.5)
plt.xlabel('Sampling Strategy', fontdict=font1)
# plt.ylabel('', fontdict=font1)
# plt.title('AUC Value',fontdict=font3)
plt.text(0.6, 0.945, r'$Mean=0.932$',fontdict=font2)
plt.text(1.5, 0.9565, 'AUC Value',fontdict=font3)
# plt.xlim(40, 160)
plt.ylim(0.90,0.96)
plt.grid(False)
plt.savefig('hist.tif', format="tif", dpi=400, bbox_inches='tight')
plt.show()
```
